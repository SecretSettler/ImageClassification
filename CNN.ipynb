```python
import numpy as np
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import torch
from torchvision import datasets, transforms
import matplotlib.pyplot as plt
```

```python
input_size = 28
num_classes = 10
num_epoches = 5
batch_size = 128

train_set = datasets.MNIST(root='./data', train=True, download=True, transform=transforms.ToTensor())

test_set = datasets.MNIST(root='./data', train=False, transform=transforms.ToTensor())

train_ds = torch.utils.data.DataLoader(dataset=train_set, batch_size=batch_size, shuffle=True)

test_ds = torch.utils.data.DataLoader(dataset=test_set, batch_size=batch_size)
```

```python
for x, y in train_ds:
    print(x.size())
```

```python
class CNN(nn.Module):
    def __init__(self):
        super(CNN, self).__init__()
        self.conv1 = nn.Sequential(
            nn.Conv2d(
                in_channels = 1,
                out_channels = 16,
                kernel_size = 5,
                stride = 1,
                padding = 2,
            ),
            nn.ReLU(),
            nn.MaxPool2d(2)
        )
        
        self.conv2 = nn.Sequential(
            nn.Conv2d(
                in_channels = 16,
                out_channels = 32,
                kernel_size = 5,
                stride = 1,
                padding = 2
            ),
            nn.ReLU(),
            nn.MaxPool2d(2)
        )
        
        self.out = nn.Linear(32 * 7 * 7, 10)
        
    def forward(self, x):
        x = self.conv1(x)
        x = self.conv2(x)
        x = x.view(x.size(0), -1)
        output = self.out(x)
        return output
```

```python
def accuracy(prediction, labels):
    pred = torch.max(prediction.data, 1)[1]
    correctness = pred.eq(labels.data.view_as(pred)).sum()
    return correctness, len(labels)
```

## Training

```python
net = CNN()
critizer = F.cross_entropy
optimizer = optim.Adam(net.parameters(), lr=0.001)
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")

net.to(device)

for epoch in range(num_epoches):
    train_rights = []
    
    for batch_i, (data, target) in enumerate(train_ds):
        net.train()
        data = data.to(device)
        target = target.to(device)
        output = net(data)
        loss = critizer(output, target)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
        right = accuracy(output, target)
        train_rights.append(right)
    
        if epoch % 100 == 0:
            net.eval()
            val_rights = []
        
            for (data, target) in test_ds:
                data = data.to(device)
                target = target.to(device)
                output = net(data)
                right = accuracy(output, target)
                val_rights.append(right)
                
            train_val = [sum(tup[0] for tup in train_rights), sum(tup[1] for tup in train_rights)]
            test_val = [sum(tup[0] for tup in val_rights), sum(tup[1] for tup in val_rights)]
            
            print('当前epoch: {} [{}/{} ({:.0f}%)]\t损失: {:.6f}\t训练集准确率: {:.2f}%\t测试集正确率: {:.2f}%'.format(
                epoch, batch_i * batch_size, len(train_ds.dataset),
                100 * batch_i / len(train_ds), 
                loss.data, 
                100. * train_val[0].cpu().numpy() / train_val[1], 
                100. * test_val[0].cpu().numpy() / test_val[1]))
```
